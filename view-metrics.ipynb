```python
import pandas as pd
import json
import os
```

```python
folder_path = "outputs/eval/metrics"
```

```python
data = {}
for filename in os.listdir(folder_path):
    if not filename.endswith(".json"):
        continue
    model_type = filename[:4]
    seed_part = filename[13:20] if "_" not in filename[13:20] else filename[13:18]
    label = f"{model_type}_{seed_part}"
    with open(os.path.join(folder_path, filename), "r") as f:
        contents = json.load(f)
    if label not in data:
        data[label] = {}
    data[label].update(contents)

data
```

```python
data_export = pd.DataFrame(data)

column_renames = {
    "ddpm_seed0": "DDPM (seed 0)",
    "ddpm_seed542": "DDPM (seed 542)",
    "flow_seed0": "Flow (seed 0)",
    "flow_seed542": "Flow (seed 542)",
}
row_renames = {
    "inception_score_mean": "IS (mean)",
    "inception_score_std": "IS (std)",
    "frechet_inception_distance": "FID",
    "kernel_inception_distance_mean": "KID (mean)",
    "kernel_inception_distance_std": "KID (std)",
}

data_export = data_export.rename(columns=column_renames, index=row_renames)
data_export = data_export[["DDPM (seed 0)", "DDPM (seed 542)", "Flow (seed 0)", "Flow (seed 542)"]]
data_export.index.name = "Metric"

data_export.to_csv("outputs/eval/metrics/metrics_summary.csv")
```
